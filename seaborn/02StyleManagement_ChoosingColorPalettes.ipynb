# Choosing Color Palettes

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={"figure.figsize":(6,6)})
np.random.seed(sum(map(ord, "palettes")))
```

# Building Color Palettes
**color_palette()**可接受seaborn和matplotlib的colormap的任意参数.  
单独调用**color_palette()**没有参数将会返回当前的默认color cycle.  
**set_palette()**设置调色板.  


# Qualitative color palettes
当import seaborn时，默认的color cycle设置为6个颜色

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

这6个主题分别为**deep**,**muted**,**pastel**,**bright**,**dark**,**colorblind**

# Using circular color systems
当超过6中颜色时间，使用**hls**颜色空间

```python
sns.palplot(sns.color_palette('hls',8))
```

**hls_palette()**函数可以控制亮度和饱和度

```python
sns.palplot(sns.hls_palette(8, l=0.3, s=0.8))
sns.palplot(sns.hls_palette(8, l=0.3, s=0.5))
sns.palplot(sns.hls_palette(8, l=0.6, s=0.5))
```

**husl**

```python
sns.palplot(sns.color_palette("husl", 8))
sns.palplot(sns.husl_palette(8, l=0.9, s=0.3))
```

# Using named colors from the xkcd color survey

# Sequential Color Palettes

```python
sns.palplot(sns.color_palette('Blues'))
sns.palplot(sns.color_palette('Greens'))
sns.palplot(sns.color_palette('Blues_r'))
```

```python
sns.palplot(sns.color_palette('GnBu_d'))
sns.palplot(sns.color_palette('GnBu'))
```

# Sequential 'cubehelix' palettes
**cubehelix**使亮度线性增加或减少  
**cubehelix_palette()**函数的**start参数(0-3)**控制颜色起始，**rot参数(-1,1)**控制旋转

```python
sns.palplot(sns.color_palette('cubehelix', 8))
```

```python
sns.palplot(sns.cubehelix_palette(8))
sns.palplot(sns.cubehelix_palette(8, start=.5, rot=-.75))
sns.palplot(sns.cubehelix_palette(8, start=2, rot=0, dark=0, light=.95, reverse=True))
```

**as_cmap**=True可以作为matplotlib的cmap

```python
x, y = np.random.multivariate_normal([0, 0],[[1, -.5],[-.5, 1]], size=300).T
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
sns.kdeplot(x, y, cmap=cmap, shade=True)
```

# Custom Sequential Palettes
**light_palette()**或者**dark_palette()**函数从dark或light生成到颜色序列

```python
sns.palplot(sns.light_palette('green'))
sns.palplot(sns.dark_palette('green'))
sns.palplot(sns.light_palette('navy'))
sns.palplot(sns.light_palette('navy', reverse=True))
```

```python
pal=sns.dark_palette('palegreen', as_cmap=True)
sns.kdeplot(x, y, cmap=pal)
```

# Setting the default color palette


```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sns.set_palette('husl')
sinplot()
```

```python
with sns.color_palette('PuBuGn_d'):
    sinplot()
```
